```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
read=pd.read_csv("C:/Users/laasy/Desktop/datasets/DelayedFlights.csv")
data=pd.DataFrame(read)
data
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
data
```

```python
data.drop('Year',axis=1,inplace=True)
data
```

```python
ace_mean=data['ActualElapsedTime'].mean()
ace_data=data['ActualElapsedTime']
ace_data.fillna(ace_mean,inplace=True)  
crse_data=data['CRSElapsedTime'] 
crse_data.fillna(crse_data.mean(),inplace=True)
air_time=data['AirTime']
air_time.fillna(air_time.mean(),inplace=True)
arr_delay_data=data['ArrDelay'] 
arr_delay_data.fillna(arr_delay_data.mean(),inplace=True)
taxin_data=data['TaxiIn'] 
taxin_data.fillna(taxin_data.mean(),inplace=True)
carrier_delay_data=data['CarrierDelay']
carrier_delay_data.fillna(carrier_delay_data.mean(),inplace=True)
weather_delay_data=data['WeatherDelay']
weather_delay_data.fillna(weather_delay_data.mean(),inplace=True)
security_delay_data=data['SecurityDelay']
security_delay_data.fillna(security_delay_data.mean(),inplace=True)
lateAir_delay_data=data['LateAircraftDelay']
lateAir_delay_data.fillna(lateAir_delay_data.mean(),inplace=True)
arr_time_data=data['ArrTime']
arr_time_data.fillna(arr_time_data.mean(),inplace=True)
data.drop(columns='TailNum',axis=1,inplace=True)
#tail_num.fillna('N325SW',inplace=True)
nasd=data['NASDelay']
nasmean=nasd.mean()
nasd.fillna(nasmean,inplace=True)
data
```
